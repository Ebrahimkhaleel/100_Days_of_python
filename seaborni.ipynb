
Seaborn is a Python data visualization library built on top of  was created by Micheal Waskom, matplotlib  provides a high-level interface for creating attractive and informative statistical graphics. It can work with DataFrames and arrays, making it particularly useful for data analysis.

```python
#as usual before using any library you have to install(pip install seaborn)
#conventionally imported as
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#loading default datasets from seaborn
tips=sns.load_dataset('tips')
tips.tail()
```

DISTRIBUTION PLOT
Distplot function in Seaborn is used to visualize the distribution of univariant data (data with a single variable) like a histogram. Now deprecated to histplot
Deprecating meaning - "Deprecated" in software development refers to a feature or functionality that is considered outdated, discouraged, and likely to be removed in future versions of the software. 

```python
#HISTPLOT vs distplot
sns.distplot(tips['tip'],kde=True)
#bins
sns.histplot(tips['tip'],kde=True,bins=30)
```

JOINT PLOT
A joint plot refers to a type of data visualization that displays the relationship between two variables in a dataset. It combines multiple plots or visual elements to provide insights into the joint distribution, correlation, and individual distributions of the variables.

```python
#kind of plot can be set [hex,reg,kde]
sns.jointplot(x='total_bill',y='tip',data=tips)
```

PAIRPLOT

A pairplot is a type of data visualization commonly used in exploratory data analysis (EDA) to visualize pairwise relationships and distributions between multiple variables in a dataset.Useful when dealing with datasets containing several numerical variables.

```python
#you can specify hue diffrentiate mostly categorical data 
#color palette can be specified
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

RUGPLOT
A rug plot is a type of univariate data visualization that is often used to display the distribution of a single variable along an axis. It consists of small vertical lines (ticks) placed at the location of each data point on the axis, resembling a "rug" of data points.

```python
#RUGPLOT
sns.rugplot(tips['total_bill'])
```

CATEGORICAL PLOTs
BARPLOT
A bar plot is a type of data visualization that represents categorical data using rectangular bars. It is commonly used to compare and display the distribution, frequency, or relationship between different categories or groups in a dataset. 

```python
#BARPLOT
#the barplot usually uses the mean as a default 
sns.barplot(x='sex',y='total_bill',data=tips)
#to add a different aggregate function we use numpy
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.sum)
```

COUNTPLOT

```python
#COUNTPLOT
#count plot is similar to barplot the basic difference is the aggregate function is (count)
#knowing the aggregate function (the y axis has been chosen) we just need the x value
sns.countplot(x='sex',data=tips)
```

BOXPLOTS(box and whisker plot) and VIOLINPLOTS

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of numerical data through quartiles. It provides a visual summary of the central tendency, dispersion, and skewness of the data along with identifying outliers.
a violin plot is a data visualization that combines elements of a box plot and a kernel density plot. It provides insights into the distribution and probability density of numerical data across different categories or groups.

```python
#boxplot

sns.boxplot(x='day',y='total_bill',data=tips)
#adding hue for multiple categories to be added to the plot
sns.boxplot(x='day',y='total_bill',data=tips,hue='time')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
#using hue to add a new category of data 
#adding split value to be true to get them in one kde
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

STRIPPLOT

```python
#STRIPPLOT (scatter plot x as the categorucal data and y as the numerical daata)
#specifying jitter True in other to get a more accurate visual of how the values are
#adding split value to be true to get them in one kde
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',orient='v')
```

#SWARMPLOT
more like combination of violinplot and 
NB:advisable not to use swarmplot for large dataset

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

CAT PLOT 
Categorical plot formerly factorplot.catplot is a powerful function in the Seaborn data visualization library that is designed specifically for creating categorical plots. It offers a high level of flexibility and customization for visualizing relationships between categorical variables in your data.


```python
#FACTORPLOT
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```

MATRIX PLOTS
Matrix plots, also known as heatmap plots, are visualizations used to display data in a matrix format where each cell's color represents the value of a variable. These plots are particularly useful for exploring relationships and patterns in large datasets.

```python
flight=sns.load_dataset('flights')
flight.tail()
```

A heatmap is a graphical representation of data where the values of a matrix are represented as colors. Heatmaps are commonly used to visualize large datasets and identify patterns or trends in the data.

```python
#HEATMAP
fc=flight.pivot_table(index='month',columns='year',values='passengers')
fc
```

```python
#Cmap which stands for color map 
#we can specify line widths and color to seperate between boxes
sns.heatmap(fc,cmap='magma',linewidths=3,linecolor='white')
```

CLUSTER MAP
clusters columns and rows based on their similarity, so labels are not in order but rather based on their similarities

```python
#CLUSTERMAP
#standard scale can be used using the standard scale parameter
sns.clustermap(fc,standard_scale=0)
```
